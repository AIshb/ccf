```python
%matplotlib inline
import time
from collections import defaultdict
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('TRAIN.csv')
test = pd.read_csv('TEST2.csv')
```

```python
train_time = train.ARRIVAL_TIME
```

```python
dic = defaultdict(int)
for x in train_time:
    dic[int(str(x)[6:8])] += 1
x = list(dic.keys())
y = list(dic.values())
plt.scatter(x, y)
```
