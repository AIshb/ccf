```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
train = pd.read_csv('TRAIN.csv')
test = pd.read_csv('TEST2.csv')
```

```python
userid = train.USERID
shopid = train.SHOPID
```

```python
user_dic = defaultdict(int)
for x in userid:
    user_dic[x] += 1
plt.bar(list(user_dic.keys()), list(user_dic.values()))
```

```python
shop_dic = defaultdict(int)
for x in shopid:
    shop_dic[x] += 1
plt.bar(list(shop_dic.keys()), list(shop_dic.values()))
```

```python
for k, v in user_dic.items():
    if v > 10:
        print(k)
```

```python
for k, v in shop_dic.items():
    if v > 20:
        print(k)
```
